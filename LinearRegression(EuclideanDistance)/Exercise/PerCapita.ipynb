```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('percapita')
plt.scatter(df.year, df.percapita,color='red' ,marker='+')
```

```python
new_df = df.drop('percapita',axis='columns')
new_df
```

```python
percapita = df.percapita
percapita
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,percapita)
```

```python
reg.predict([[2020]]) #Ans 41288.69409442 
```

```python
reg.coef_ # slope or m
```

```python
reg.intercept_
```

```python
#y=mx+b
828.46507522*2020-1632210.7578554575
```

```python
area_df = pd.read_csv("year.csv")
area_df
```

```python
p = reg.predict(area_df)
p
```

```python
area_df['percapita']=p
area_df
```
