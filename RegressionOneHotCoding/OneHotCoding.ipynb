```python
import pandas as pd
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
pd.concat([df, dummies],axis='columns')
```

```python
df_dummies = pd.concat([df, dummies],axis='columns')
df_dummies
```

```python
df_dummies.drop(['town','west windsor'],axis='columns' , inplace=True)
df_dummies # 2 coulmns dropped due to dummy variable trap to find the derivation between two 2 columns (monroe township,robinsville) and maintain the collinearity between 2 
```

```python
X = df_dummies.drop('price',axis='columns')
X
```

```python
y = df_dummies.price
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
model.predict([[3400,0,0]])
```

```python
model.predict([[2800,0,1]])
```

```python
df = pd.read_csv("carprices.csv")
df
```

```python
dummies = pd.get_dummies(df.CarModel)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['CarModel'], axis='columns')
final
```

```python
final = final.drop(['Mercedez Benz C class'], axis='columns')
final
```

```python
X = final.drop('SellPrice', axis='columns')
X
```

```python
y = final.SellPrice
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict(X)
```

```python
model.score(X,y)
```
